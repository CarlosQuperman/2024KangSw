```python
#색상값 측정 예제

from spike import ColorSensor
from spike.control import wait_for_seconds

# 컬러 센서를 초기화합니다.
color_s = ColorSensor('E')
color = color_s.get_color()

while color!='red':
    color = color_s.get_color()
    print(color)
    if(color=='red'):
        print('find red')
    wait_for_seconds(1)
```

```python
#반사광 측정 예제

from spike import ColorSensor
from spike.control import wait_for_seconds

color_sensor = ColorSensor('E')
light = color_sensor.get_reflected_light()

while light!=0:
    light = color_sensor.get_ambient_light()
    print(light)
    wait_for_seconds(1)
```

```python
#지정 각도로 회전하기

from spike import Motor
from spike.control import wait_for_seconds

motor = Motor('F')  #모터 초기화

motor.run_to_position(90) #90도 위치로 모터 회전

for deg in range(0, 361, 90): #0도부터 360도까지 1초 간격으로 90도씩 회전하기
    motor.run_to_degrees_counted(deg)
    wait_for_seconds(1)

motor.run_for_degrees(90) #90도 회전하기
```
