```python
#빛(컬러) 센서에 따라 전진 후진 운동을 반복하게 하는 프로그램

from spike import PrimeHub, Button, ColorSensor, App, Motor, MotorPair
from spike.control import wait_for_seconds, wait_until, Timer
from math import *
from spike import Motor
from spike import App

#hub 초기화
hub = PrimeHub()

#모터 및 컬러 센서 설정 및 이름 설정
motor_pair = MotorPair('A', 'B')
color = ColorSensor('C')


def go_back_20():
  while not hub.right_button.is_pressed():
      motor_pair.start_tank(-20,-20)

      if color.get_reflected_light()>10:
          motor_pair.start_tank(20,20)
          wait_for_seconds(3)


go_back_20()
```
